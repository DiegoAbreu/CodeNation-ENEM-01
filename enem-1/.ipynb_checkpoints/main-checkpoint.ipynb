### Importando o CSV

```python
import csv
```

```python
import pandas as pd
data = "train.csv"
df = pd.read_csv(data)
```

```python
df
```

### Separando apenas as colunas necessárias

```python
enem = pd.DataFrame(df, columns=['NU_INSCRICAO', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO'])
```

```python
enem.head(10)
```

### Eliminando os valores Nulos

```python
enem = enem.dropna(axis=0, how='any')
```

```python
enem.head(10)
```

### Calculando as médias

```python
enem['MEDIA']= ((enem['NU_NOTA_MT'] * 3) + (enem['NU_NOTA_CN'] * 2) + (enem['NU_NOTA_LC'] * 1.5) + (enem['NU_NOTA_CH'] * 1) + (enem['NU_NOTA_REDACAO'] * 3))/5
```

```python
enem.head(10)
```

### Criando um dataframe com o resultado

```python
resultado = pd.DataFrame(enem, columns=['NU_INSCRICAO', 'MEDIA'])
```

```python
resultado.head()
```

```python
resultado = resultado.sort_values(['MEDIA'], ascending=False)
```

```python
resultado = resultado.head(20)
```

```python
resultado
```

### Exportando a Resposta em arquivo .CSV

```python
resultado.to_csv('answer.csv', index=False)
```
